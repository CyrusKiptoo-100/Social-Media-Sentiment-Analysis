```python
sme.isnull().sum()
```

```python
#Convert posttime into datetime format
sme['post_time']=pd.to_datetime(sme['post_time'])
```

```python
sme['post_hour']=sme['post_time'].dt.hour
sme['post_day']=sme['post_time'].dt.day
```

*FEATURE ENGINEERING*

```python
# create weekend post and weekday post
sme['weekend_post']=sme['post_day'].isin(['Saturday','Sunday']).astype(int)
sme['weekday_post']=sme['weekend_post'].astype(int)
```

```python
#Calculate caption_lenght
sme["caption_length"]=sme["caption"].str.len()
```

DATA VISUALIZATION

```python

#Compare average engagement across platforms
sns.barplot(x='platform', y='likes', data=sme)
plt.show()
```

```python
# Correlation matrix
correlation_matrix = sme[['likes', 'comments', 'shares', 'caption_length', 'num_hashtags']].corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

POSITIVE CORRELATION:
comment and Shares suggests that posts with more likes tend to have more shares.
It might indicate porpular content measured by likes is more likely to be shared to Users.

NEGATIVE CORR.
Caption lentgh and engagements
suggests that post with long caption lenghts tend to have low number engagements resulting to low shares.
This shows that users prefer short and more concise contents
